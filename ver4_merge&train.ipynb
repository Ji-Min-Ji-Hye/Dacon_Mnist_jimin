# cnn_model

```python
def create_cnn_model(x_train):  #layer추가
    inputs = tf.keras.layers.Input(x_train.shape[1:]) #x_train.shape[1:]=(28,28,1)
    
    #배치 정규화는 활성화함수 통과 전, 가중의 변화를 줄이는 것이 목표
    bn = tf.keras.layers.BatchNormalization()(inputs)
    
    conv = tf.keras.layers.Conv2D(8, kernel_size=3, strides=1, padding='same', activation='relu')(bn)
    bn = tf.keras.layers.BatchNormalization()(conv)
    conv = tf.keras.layers.Conv2D(8, kernel_size=2, strides=1, padding='same', activation='relu')(bn)
    pool = tf.keras.layers.MaxPooling2D((2, 2))(conv)
    
    conv = tf.keras.layers.Conv2D(16, kernel_size=3, strides=1, padding='same', activation='relu')(bn)
    bn = tf.keras.layers.BatchNormalization()(conv)
    conv = tf.keras.layers.Conv2D(16, kernel_size=2, strides=1, padding='same', activation='relu')(bn)
    pool = tf.keras.layers.MaxPooling2D((2, 2))(conv)
    
    conv = tf.keras.layers.Conv2D(32, kernel_size=2, strides=1, padding='same', activation='relu')(bn)
    bn = tf.keras.layers.BatchNormalization()(conv)
    conv = tf.keras.layers.Conv2D(32, kernel_size=2, strides=1, padding='same', activation='relu')(bn)
    pool = tf.keras.layers.MaxPooling2D((2, 2))(conv)
    
    conv = tf.keras.layers.Conv2D(64, kernel_size=2, strides=1, padding='same', activation='relu')(bn)
    bn = tf.keras.layers.BatchNormalization()(conv)
    conv = tf.keras.layers.Conv2D(64, kernel_size=2, strides=1, padding='same', activation='relu')(bn)
    pool = tf.keras.layers.MaxPooling2D((2, 2))(conv)

    #stride 크기만큼 이미지 이동
    conv = tf.keras.layers.Conv2D(128, kernel_size=3, strides=1, padding='same', activation='relu')(bn)
    #출력결과: 1+(입력크기-필터크기)/stride 크기= 1+(28-5)/1=24
    
    bn = tf.keras.layers.BatchNormalization()(conv) 
    conv = tf.keras.layers.Conv2D(128, kernel_size=2, strides=1, padding='same', activation='relu')(bn)
    pool = tf.keras.layers.MaxPooling2D((2, 2))(conv)
    bn = tf.keras.layers.BatchNormalization()(pool)
    conv = tf.keras.layers.Conv2D(256, kernel_size=2, strides=1, padding='same', activation='relu')(bn)
    bn = tf.keras.layers.BatchNormalization()(conv)
    conv = tf.keras.layers.Conv2D(256, kernel_size=2, strides=1, padding='same', activation='relu')(bn)
    pool = tf.keras.layers.MaxPooling2D((2, 2))(conv)

    flatten = tf.keras.layers.Flatten()(pool)

    bn = tf.keras.layers.BatchNormalization()(flatten)
    #dense: Dense 레이어는 입력과 출력을 모두 연결해줌. relu와 주로 같이 쓰임.
    dense = tf.keras.layers.Dense(1000, activation='relu')(bn)
    
    bn = tf.keras.layers.BatchNormalization()(dense)
    
    #softmax: 가장 큰 출력값을 부여받은 클래스가 확률이 가장 높은것으로 이용
    outputs = tf.keras.layers.Dense(10, activation='softmax')(bn)
    
    model = tf.keras.models.Model(inputs=inputs, outputs=outputs)

    return model
```

```python
model = create_cnn_model(x_train)  #summary보고 이해하기
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
x_train.shape
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
earlystopper = EarlyStopping(patience=7, monitor='val_loss')    #성능이 더이상 증가하지 않을 때 학습을 중지함. 
best_weights = ModelCheckpoint('best.h5', verbose=1, monitor='val_loss', save_best_only=True, mode='auto')

model.fit(x_train, y_train, epochs=200,batch_size=32,callbacks=[earlystopper, best_weights])
```
